```
# Starting with importing the usefull packages.
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```
# Load the iris dataset
iris = pd.read_csv('/content/Iris.csv')
iris.head() # See the first 5 rows
```

```
label = LabelEncoder()
iris['Species'] = label.fit_transform(iris['Species'])
iris.head(10)
```

```
iris.columns
```

```
colsname  = list(iris.columns)
predictors = colsname[1:5]
target = colsname[5]
```

```
#Swarmplot
sns.swarmplot(x = 'Species', y = 'SepalLengthCm', data = iris)
```

```
sns.swarmplot(x = 'Species', y = 'PetalLengthCm', data = iris)
```

```
train, test = train_test_split(iris, test_size = 0.25)
```

```
model = DecisionTreeClassifier(criterion = "entropy")
model.fit(train[predictors],train[target])
```

```
preds = model.predict(test[predictors])
np.mean(preds==test.Species) 
#Accuracy
```

```
pd.crosstab(test[target],preds)
```

```
plt.figure(figsize=(25,10))
a = plot_tree(model, 
              feature_names=predictors, 
              class_names=target, 
              filled=True, 
              rounded=True, 
              fontsize=14)
```
